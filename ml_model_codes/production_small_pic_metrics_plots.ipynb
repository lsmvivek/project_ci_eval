```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
which_plot = ''
# which_plot = '_training'
save_plot = False
target_ = 'SoilMoist_S_tavg'     #  Tveg_tavg  Qs_tavg  SoilMoist_S_tavg
dir_ = '82.875_27.625'  # Ganga basin
# dir_ = '148.125_-29.125'      # Murray basin
```

```python
# pcc_mcc_df = pd.read_csv('data/temporary/PCC-MC-results_{target_}_{dir_}.txt', index_col=None)
pcc_mcc_df = pd.read_csv(f'data/temporary/PCC-MC-results_{target_}_{dir_}{which_plot}.txt', index_col=None)
pcc_mcc_df
```

```python
# tcdf_mcc_df = pd.read_csv('data/temporary/TCDF-MC-results_{target_}_{dir_}.txt', index_col=None)
tcdf_mcc_df = pd.read_csv(f'data/temporary/TCDF-MC-results_{target_}_{dir_}{which_plot}.txt', index_col=None)
tcdf_mcc_df
```

```python
# varlingam_mcc_df = pd.read_csv('data/temporary/VARLiNGAM-MC-results_{target_}_{dir_}.txt', index_col=None)
varlingam_mcc_df = pd.read_csv(f'data/temporary/VARLiNGAM-MC-results_{target_}_{dir_}{which_plot}.txt', index_col=None)
varlingam_mcc_df
```

```python
# pcmciplus_mcc_df = pd.read_csv('data/temporary/PCMCIplus-MC-results_{target_}_{dir_}.txt', index_col=None)
pcmciplus_mcc_df = pd.read_csv(f'data/temporary/PCMCIplus-MC-results_{target_}_{dir_}{which_plot}.txt', index_col=None)
pcmciplus_mcc_df
```

```python
# dynotears_mcc_df = pd.read_csv('data/temporary/DYNOTEARS-MC-results_{target_}_{dir_}.txt', index_col=None)
dynotears_mcc_df = pd.read_csv(f'data/temporary/DYNOTEARS-MC-results_{target_}_{dir_}{which_plot}.txt', index_col=None)
dynotears_mcc_df
```

```python
# # Reorder the dataframe based on the methods
# # TCDF, PCMCIplus, DYNOTEARS, VARLiNGAM, PCC
# method_order = ['TCDF', 'PCMCIplus', 'DYNOTEARS', 'VARLiNGAM', 'PCC']
# monte_carlo_df['method'] = pd.Categorical(monte_carlo_df['method'], categories=method_order, ordered=True)
```

```python
monte_carlo_df = pd.concat([pcc_mcc_df, tcdf_mcc_df, varlingam_mcc_df, pcmciplus_mcc_df, dynotears_mcc_df], ignore_index=True)

monte_carlo_df
```

```python
# plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.family"] = "sans-serif"

sns.set_theme(context='notebook', style='white', font='sans-serif', font_scale=1,)
fig, ax = plt.subplots(1, 1, figsize=(5, 5), dpi=100)

sns.boxplot(
    data=monte_carlo_df, x='noise_level', y='r2_score', hue='method',
    orient='v', width=1, dodge=True, ax=ax, legend=False, fill=False,
    flierprops={'marker': 'o', 'markersize': 0.3},
    palette=sns.color_palette('colorblind', 5),
    # palette=custom_palette,
    whis=[5, 90], linewidth=1.,
    )
noise_levels = monte_carlo_df['noise_level'].unique()
for i, nl in enumerate(noise_levels):
    if i % 2 == 0:
        ax.axvspan(i - 0.5, i + 0.5, facecolor="lightgrey", alpha=0.2, zorder=0)
        ax.axvspan(i - 0.5, i + 0.5, facecolor="lightgrey", alpha=0.2, zorder=0)
ax.set_ylim(0, 1.0)
# ax.grid(True, linestyle='-', linewidth=0.5, alpha=0.5)
ax.yaxis.set_label_coords(-0.1, 0.5)
ax.grid(True, linestyle="-", linewidth=0.5, color='grey', alpha=0.3, zorder=-1, axis='y')
ax.set_xlabel('Noise Level', fontsize=12)
ax.set_ylabel(r'$\mathrm{R^2}$', fontsize=12)
ax.tick_params(axis='y', labelsize=12)
ax.tick_params(length=5, axis='y', which='major', direction='out', bottom=True, left=True)
plt.yticks(fontsize=12)
ax.minorticks_on()
ax.tick_params(length=2, axis='y', which='minor', direction='out', bottom=True, left=True)
# Add a boundary around the scatterplot
for spine in ax.spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(0.5)
    spine.set_alpha(0.2)
plt.tight_layout()
if save_plot:
    plt.savefig(f'./figs/{which_plot}_r2.svg', bbox_inches='tight', dpi=400, pad_inches=0.05)
plt.show()
```
