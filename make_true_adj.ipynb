##### Read true adjacency matrix created from literature

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib 
import seaborn as sns
import os
import sys

sys.path.append('./ci_eval/python_scripts/')
os.chdir('./ci_eval/')
from create_adj_mat_figure import create_adj_mat_figure
# import proplot
```

```python
temp=pd.read_excel('data/adjacency_matrix/clsm_tru_directional_working.xlsx', sheet_name=1, index_col=0).drop(columns=['Unnamed: 1'])
temp
```

```python
# Create a new DataFrame with the same columns and new row names (original row name with "lag1" suffix)
lag1_df = pd.DataFrame(0, index=[f"{idx}_lag1" for idx in temp.index], columns=temp.columns)

# Copy the diagonal values from temp to lag1_df
for i in range(len(temp)):
    lag1_df.iat[i, i] = temp.iat[i, i]
    temp.iat[i, i]=0
# Join the original temp and the lag1_df vertically
combined_df = pd.concat([temp, lag1_df], axis=0)
# combined_df.to_csv('data/clsm_tru_directional_working_lag1.csv', sep=',')
combined_df
```

```python
combined_df = combined_df.transpose()
sns.heatmap(combined_df, cmap='viridis')
```

```python
# New dictionary containing key value pairs
# keys are the effects and values are the causes
# Later we iterate through this dictionary to create the adjacency matrix
new_combined_df = combined_df.copy()
new_combined_df.loc[:] = 0
new_effect_causes_dict = {'CanopInt_tavg': ['ECanop_tavg', 'Rainf_tavg*', 'Rainf_f_tavg*']}
new_effect_causes_dict['Qs_tavg'] = ['CanopInt_tavg', 'SoilMoist_S_tavg', 'Rainf_tavg*', 'Rainf_f_tavg*']
new_effect_causes_dict['SoilMoist_S_tavg'] = ['Rainf_tavg*', 'Rainf_f_tavg*', 'SoilMoist_RZ_tavg', 'SoilMoist_P_tavg',
                                              'CanopInt_tavg', 'Evap_tavg', 'ESoil_tavg', 'SoilMoist_S_tavg_lag1']
new_effect_causes_dict['SoilMoist_RZ_tavg'] = ['SoilMoist_S_tavg', 'SoilMoist_P_tavg', 'GWS_tavg', 'Evap_tavg', 
                                               'ESoil_tavg', 'SoilMoist_RZ_tavg_lag1']
new_effect_causes_dict['SoilMoist_P_tavg'] = ['SoilMoist_RZ_tavg', 'GWS_tavg', 'SoilMoist_P_tavg_lag1']
new_effect_causes_dict['GWS_tavg'] = ['SoilMoist_RZ_tavg', 'Qsb_tavg', 'GWS_tavg_lag1']
new_effect_causes_dict['Qsb_tavg'] = ['SoilMoist_P_tavg', 'GWS_tavg']
# new_effect_causes_dict['Tws_tavg'] = ['CanopInt_tavg', 'SoilMoist_S_tavg', 'SoilMoist_RZ_tavg', 'GWS_tavg', 
#                                       'SoilMoist_P_tavg', 'Tws_tavg_lag1']
new_effect_causes_dict['Evap_tavg'] = ['ACond_tavg', 'AvgSurfT_tavg', 'Tair_f_tavg*', 'Qair_f_tavg*', 'Psurf_f_tavg*',
                                       'CanopInt_tavg', 'SoilMoist_S_tavg', 'SoilMoist_P_tavg', 'Wind_f_tavg*']
new_effect_causes_dict['ECanop_tavg'] = ['Evap_tavg', 'CanopInt_tavg', 'SoilMoist_S_tavg', 'Wind_f_tavg*']
new_effect_causes_dict['ESoil_tavg'] = ['Evap_tavg', 'ECanop_tavg', 'CanopInt_tavg', 'SoilMoist_S_tavg', 'SoilMoist_RZ_tavg', 'Wind_f_tavg*']
new_effect_causes_dict['Tveg_tavg'] = ['Evap_tavg', 'ECanop_tavg', 'CanopInt_tavg', 'SoilMoist_S_tavg', 'Wind_f_tavg*']
# new_effect_causes_dict['ACond_tavg'] = ['Tair_f_tavg*', 'Qair_f_tavg*', 'Psurf_f_tavg*', 'Wind_f_tavg*']
new_effect_causes_dict['AvgSurfT_tavg'] = ['Lwdown_f_tavg*', 'Swdown_f_tavg*', 'Qh_tavg', 'Qg_tavg', 'Evap_tavg',
                                            'Tair_f_tavg*', 'Psurf_f_tavg*', 'Wind_f_tavg*']
new_effect_causes_dict['Qle_tavg'] = ['Evap_tavg', 'ECanop_tavg', 'Tveg_tavg']
new_effect_causes_dict['Qh_tavg'] = ['AvgSurfT_tavg', 'Qh_tavg_lag1']
new_effect_causes_dict['Qg_tavg'] = ['AvgSurfT_tavg', 'Qg_tavg_lag1']
new_effect_causes_dict['Lwnet_tavg'] = ['Lwdown_f_tavg*', 'AvgSurfT_tavg']
new_effect_causes_dict['Swnet_tavg'] = ['Swdown_f_tavg*', 'AvgSurfT_tavg']


new_effect_causes_dict['Tws_tavg'] = ['CanopInt_tavg', 'GWS_tavg', 'SoilMoist_RZ_tavg']
new_effect_causes_dict['ACond_tavg'] = ['AvgSurfT_tavg',
                                        'Tair_f_tavg*', 'Qair_f_tavg*', 'Psurf_f_tavg*', 'Wind_f_tavg*',
                                        'CanopInt_tavg', 'SoilMoist_S_tavg',]
for key, value in new_effect_causes_dict.items():
    print(f"{key}: ", end='\n')
    for v in value:
        print(v, end=', ')
        print(combined_df.loc[key, v], end=', ')
        # Set value to 1
        new_combined_df.loc[key, v] = 1
    print('\n')
```
